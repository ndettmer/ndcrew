```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
C = 1.
g_Na = 120.
g_K = 36.
g_L = .3
E_Na = 50.
E_K = -77.
E_L = -54.5
I_ext = 10

# TODO: as funciton for connections!
I_syn = 0.

tau_d = 3.
tau_r = .1
N = 500.
```

```python
def alpha_m(V):
    """S5"""
    return .1 * (V + 40.) / (1. - np.exp(-(V + 40.) / 10.))

def beta_m(V):
    """S6"""
    return 4. * np.exp(-(V + 65.) / 18.)

def alpha_h(V):
    """S7"""
    return .07 * np.exp(-(V + 65.) / 20.)

def beta_h(V):
    """S8"""
    return 1. / (1. + np.exp(-(V + 35.) / 10.))

def alpha_n(V):
    """S9"""
    return ((V + 55.) / 100.) / (1. - np.exp(-.1 * (V + 55.))) #/100 statt 10

def beta_n(V):
    """S10"""
    return .125 * np.exp(-(V + 65.) / 80.)
    
def dm(V, m):
    """S2"""
    a = alpha_m(V)
    b = beta_m(V)
    
    return a * (1. - m) - b * m

def dh(V, h):
    """S3"""
    a = alpha_h(V)
    b = beta_h(V)
    
    return a * (1. - h) - b * h

def dn(V, n):
    """S4"""
    a = alpha_n(V)
    b = beta_n(V)
    
    return a * (1. - n) - b * n

def dV(V, m, h, n):
    """S1"""
    return (- g_Na * m ** 3 * h * (V - E_Na) 
            - g_K * n ** 4 * (V - E_K) 
            - g_L * (V - E_L) 
            + I_ext + I_syn ) / C
```
